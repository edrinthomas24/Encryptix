```python
#correlation matrix 
# thw value willl range from -1 to 1 
columns=['sepal_length','sepal_width','petal_length','petal_width','species']
iris = pd.read_csv(r'C:\Users\Edrin Thomas\Downloads\IRIS.csv', header=None, names=columns)
#numerical_columns = iris.select_dtypes(include=['float64', 'int64'])
print("First few rows of the dataset:")
print(iris.head())

# Ensure there are no header misinterpretations
iris = iris[pd.to_numeric(iris['sepal_length'], errors='coerce').notnull()]

# Convert appropriate columns to numerical types
iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']] = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].astype(float)

# Check the data types after conversion
print("\nData types after conversion:")
print(iris.dtypes)

# Select only the numerical columns
numerical_columns = iris.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation
correlation_matrix = numerical_columns.corr()

print("\nCorrelation matrix:")
print(correlation_matrix)
```

```python
 #Label Encoding 
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
# convert the species into numeric Data types .
```

```python
df['species'] = le.fit_transform(df['species'])
df.head()
```

```python

#Model Training 
from sklearn.model_selection import train_test_split
x=df.drop(columns=['species'])
y=df['species']
x_train,x_test,y_train,y_test = train_test_split(x,y ,test_size=0.20)
```

```python
#logistic Regression  MODEL -1 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#print matrix to get performance 
print("Accuracy :",model.score(x_test,y_test)*100,"%")
```

```python

# knn - MODEL -2 
from sklearn.neighbors import KNeighborsClassifier
model =KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
# acurray of knn model 
print("Accuracy :",model.score(x_test,y_test)*100,"%")
```

```python
prediction = model.predict(x_test)
from sklearn.metrics import accuracy_score 
print(accuracy_score(y_test, prediction)*100)
```
